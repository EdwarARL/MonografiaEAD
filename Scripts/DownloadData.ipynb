Se descargarán los datos para una ventana de 6 años, desde el 01/01/2013 hasta el 31/12/2018.

#### Descarga de la demanda del sistema por hora y análisis estadístico

```python

df_DemReal= apiXM.request_data(pydataxm.ReadDB(),    #Se indica el objeto que contiene el serivicio
                        "DemaReal",                #Se indica el nombre de la métrica tal como se llama en el campo metricID
                        "Sistema",                 #Campo que indica si es Sistema, Agente, Recurso, Comercializador
                        FechaIni,       #Corresponde a la fecha inicial de la consulta
                        FechaFin)      #Corresponde a la fecha final de la consulta
```

```python
df_DemReal.info()
```

```python
df_DemReal.head(5)
```

```python
df_DemReal['Year']=df_DemReal['Date'].dt.year
df_DemReal['Month']=df_DemReal['Date'].dt.month
df_DemReal['Day']=df_DemReal['Date'].dt.day
```

```python
df_DemReal.head(3)
```

```python
columns_name=list(df_DemReal.columns)[2:26]
```

```python
icont=1
for iname in columns_name:
    df_DemReal.rename(columns={iname:icont},inplace=True)
    icont+=1
```

```python
df_DemReal.head(3)
```

```python
columns_name=list(df_DemReal.columns)[2:26]
index_name=list(df_DemReal.columns)[26::]
```

```python
df_unpivot_dem = pd.melt(df_DemReal, id_vars=index_name, value_vars=columns_name,var_name='Hour', value_name='Demanda')
df_unpivot_dem.info()
```

```python
df_unpivot_dem.head(10)
```

```python
df_unpivot_dem.to_csv('DemandaReal_hora.csv')
```

##### Agrupar los datos de demanda por día

```python
group_dem_day=df_unpivot_dem.groupby('Date').mean()
```

```python
group_dem_day.info()
```

```python
group_dem_day.head(5)
```

#### Descarga del volumen útil en los embalses para el sistema por hora

```python
# Unit kWh
df_VolumenU= apiXM.request_data(pydataxm.ReadDB(),    #Se indica el objeto que contiene el serivicio
                        "VoluUtilDiarEner",                #Se indica el nombre de la métrica tal como se llama en el campo metricID
                        "Sistema",                 #Campo que indica si es Sistema, Agente, Recurso, Comercializador
                        FechaIni,       #Corresponde a la fecha inicial de la consulta
                        FechaFin)      #Corresponde a la fecha final de la consulta
df_VolumenU.rename(columns={'Value':'VolUtil'},inplace=True)
df_VolumenU.head(5)   
```

```python
df_VolumenU.info()
```

```python
# Agregar el volumen útil
df_DataModel = group_dem_day.merge(df_VolumenU, on='Date', how='inner')[['Date', 'Demanda', 'VolUtil']]
```

```python
df_DataModel.info(5)
```

#### Descarga de lo aportes en los caudales de los rios

```python
# Unit kWh
df_Aportes= apiXM.request_data(pydataxm.ReadDB(),    #Se indica el objeto que contiene el serivicio
                        "AporEner",                #Se indica el nombre de la métrica tal como se llama en el campo metricID
                        "Sistema",                 #Campo que indica si es Sistema, Agente, Recurso, Comercializador
                        FechaIni,       #Corresponde a la fecha inicial de la consulta
                        FechaFin)      #Corresponde a la fecha final de la consulta
df_Aportes.rename(columns={'Value':'Aportes'},inplace=True)
df_Aportes.head(5)   
```

#### Descarga del precio de bolsa nacional

```python
df_PrecioB['Year']=df_PrecioB['Date'].dt.year
df_PrecioB['Month']=df_PrecioB['Date'].dt.month
df_PrecioB['Day']=df_PrecioB['Date'].dt.day
```

```python
columns_name=list(df_PrecioB.columns)[2:26]
```

```python
icont=1
for iname in columns_name:
    df_PrecioB.rename(columns={iname:icont},inplace=True)
    icont+=1
```

#### Descarga de la oferta de los recursos

```python
# Unit COP/kWh
df_PrecioOfe= apiXM.request_data(pydataxm.ReadDB(),    #Se indica el objeto que contiene el serivicio
                        "PrecOferIdeal",                #Se indica el nombre de la métrica tal como se llama en el campo metricID
                        "Recurso",                 #Campo que indica si es Sistema, Agente, Recurso, Comercializador
                        FechaIni,       #Corresponde a la fecha inicial de la consulta
                        FechaFin)      #Corresponde a la fecha final de la consulta

df_PrecioOfe.head(5)   
```

#### Descarga de la disponibilidad

```python
# Units kWh, Descarga de la disponibilidad real
df_Disp= apiXM.request_data(pydataxm.ReadDB(),    #Se indica el objeto que contiene el serivicio
                        "DispoReal",                #Se indica el nombre de la métrica tal como se llama en el campo metricID
                        "Recurso",                 #Campo que indica si es Sistema, Agente, Recurso, Comercializador
                        FechaIni,       #Corresponde a la fecha inicial de la consulta
                        FechaFin)      #Corresponde a la fecha final de la consulta

df_Disp.head(5)   
```

```python
df_Disp.info()
```

```python
df_Disp.fillna(0, inplace=True)
df_Disp.info()
```

```python
df_Disp.to_csv('Disp.csv')
```

```python
df_Disp.head(5)
```

#### Obtener información de recursos

```python
df_Rec= apiXM.request_data(pydataxm.ReadDB(),    #Se indica el objeto que contiene el serivicio
                        "ListadoRecursos",                #Se indica el nombre de la métrica tal como se llama en el campo metricID
                        "Sistema",                 #Campo que indica si es Sistema, Agente, Recurso, Comercializador
                        FechaIni,       #Corresponde a la fecha inicial de la consulta
                        FechaFin)      #Corresponde a la fecha final de la consulta
```

```python
df_Rec.info()
```

```python
df_Rec.head(10)
```

```python
df_Disp_Type=df_Disp.merge(df_Rec,left_on=['Values_code'],right_on=['Values_Code'], how='left')
```

```python
df_Disp_Type.info()
```

#### Procesamiento de la disponibilidad de los recursos térmicos

```python
df_Termica=df_Disp_Type[df_Disp_Type['Values_Type']=='TERMICA']
```

```python
df_Termica.info()
```

```python
group_Dis_Ter_hour=df_Termica.groupby('Date_x').sum()
group_Dis_Ter_hour.reset_index(inplace=True)
group_Dis_Ter_hour.head(5)
```

```python
group_Dis_Ter_hour.rename(columns={'Date_x':'Date'},inplace=True)

group_Dis_Ter_hour['Year']=group_Dis_Ter_hour['Date'].dt.year
group_Dis_Ter_hour['Month']=group_Dis_Ter_hour['Date'].dt.month
group_Dis_Ter_hour['Day']=group_Dis_Ter_hour['Date'].dt.day

columns_name=list(group_Dis_Ter_hour.columns)[1:25]
icont=1
for iname in columns_name:
    group_Dis_Ter_hour.rename(columns={iname:icont},inplace=True)
    icont+=1

columns_name=list(group_Dis_Ter_hour.columns)[1:25]
index_name=list(group_Dis_Ter_hour.columns)[25::]
index_name.insert(0, list(group_Dis_Ter_hour.columns)[0])

df_unpivot_Disp_Ter = pd.melt(group_Dis_Ter_hour, id_vars=index_name, value_vars=columns_name,var_name='Hour', value_name='DispTer')
```

```python
df_unpivot_Disp_Ter.to_csv('Disp_Ter__Hora_agregado.csv')
```

```python
group_disp_Ter_day=df_unpivot_Disp_Ter.groupby('Date').mean()
group_disp_Ter_day
```

```python
group_disp_Ter_day.info()
```

```python
# Agregar disponibilidad real
df_DataModel = df_DataModel.merge(group_disp_Ter_day, on='Date', how='inner')[['Date', 'Demanda', 'VolUtil','Aportes','PrecioB','PrecioO','DispTer']]
df_DataModel
```

#### Procesamiento de la disponibilidad de los recursos no térmicos

```python
df_NoTermica=df_Disp_Type[df_Disp_Type['Values_Type']!='TERMICA']
```

```python
df_NoTermica.info()
```

```python
group_Dis_NoTer_hour=df_NoTermica.groupby('Date_x').sum()
group_Dis_NoTer_hour.reset_index(inplace=True)
group_Dis_NoTer_hour.head(5)
```

```python
group_Dis_NoTer_hour.rename(columns={'Date_x':'Date'},inplace=True)

group_Dis_NoTer_hour['Year']=group_Dis_NoTer_hour['Date'].dt.year
group_Dis_NoTer_hour['Month']=group_Dis_Ter_hour['Date'].dt.month
group_Dis_NoTer_hour['Day']=group_Dis_NoTer_hour['Date'].dt.day

columns_name=list(group_Dis_NoTer_hour.columns)[1:25]
icont=1
for iname in columns_name:
    group_Dis_NoTer_hour.rename(columns={iname:icont},inplace=True)
    icont+=1

columns_name=list(group_Dis_NoTer_hour.columns)[1:25]
index_name=list(group_Dis_NoTer_hour.columns)[25::]
index_name.insert(0, list(group_Dis_NoTer_hour.columns)[0])

df_unpivot_Disp_NoTer = pd.melt(group_Dis_NoTer_hour, id_vars=index_name, value_vars=columns_name,var_name='Hour', value_name='DispNoTer')
```

```python
df_unpivot_Disp_NoTer.to_csv('Disp_NoTer__Hora_agregado.csv')
```

```python
group_disp_NoTer_day=df_unpivot_Disp_NoTer.groupby('Date').mean()
group_disp_NoTer_day
```

```python
# Agregar disponibilidad real
df_DataModel = df_DataModel.merge(group_disp_NoTer_day, on='Date', how='inner')[['Date', 'Demanda', 'VolUtil','Aportes','PrecioB','PrecioO','DispTer','DispNoTer']]
df_DataModel
```
